## DBSCAN Clustering

DBSCAN is another clustering algorithm that is based on density. The algorithm seeks to generate clusters by grouping points that are densely packed together, separated by sparse regions. One distinction with DBSCAN is that it is also able to handle noise - those points that are not part of any cluster.

### DBSCAN Setup and Algorithm

Using DBSCAN is mostly the same as we are used to with other clustering algorithms, and other models in general. 

We have a couple of hyperparameters to setup that are unique to DBSCAN. 
<ul>
<li> Epsilon value, which is the maximum distance between two points for them to be considered neighbors. 
<li> Minimum number of points (minPts) that must be in a neighborhood for a point to be considered a core point.
</ul>

The algorithm works as follows:
<ol>
<li> For each point, find the number of points within epsilon distance. If the number of points is less than the minimum number of points, the point is considered noise. Otherwise, the point is a core point.
<li> For each core point, find all points within epsilon distance. If any of those points are core points, add them to the cluster. If any of those points are not core points, add them to the cluster. These are border points. 
<li> Repeat step 2 until no more points are added to the cluster.
<li> Repeat steps 1-3 for each point that is not part of a cluster.
</ol>

Roughly, if a point has "enough" neighbors (as defined by the minPts) then it is a core point. If a point is a core point, then all points within epsilon distance are part of the same cluster. If a point is not a core point, then it is part of a cluster if it is within epsilon distance of a core point. Those points that aren't within a distance to be grouped into any cluster are labeled as noise. For tuning, increasing the epsilon or decreasing the minimum points will allow the clusters to be less tightly grouped. 

![Points](images/points.webp "Points")

Note that feature scaling is important for getting reliable results with DBSCAN. If we don't scale the data, then the distance between points will be dominated by the features with the largest range.

```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np

#Determine centroids
centers = [[0.5, 2], [-1, -1], [1.5, -1]]
#Create dataset
X, y = make_blobs(n_samples=400, centers=centers, 
                  cluster_std=0.5, random_state=0)
#Normalize the values
X = StandardScaler().fit_transform(X)

plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='Paired')
```

Now we can create the DBSCAN model and plot the results. 

```python
db = DBSCAN(eps=0.2, min_samples=10)
db.fit(X)

y_pred = db.fit_predict(X)
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1],c=y_pred, cmap='Paired')
plt.title("Clusters determined by DBSCAN")
```

#### Estimating Epsilon

We can estimate a good epsilon value using the nearest neighbor distance. The nearest neighbor distance is the distance to the nearest point in the dataset. We can use the kneighbors() function to find the nearest neighbors for each point. The kneighbors() function returns the distances to the nearest neighbors, as well as the indices of the nearest neighbors. We can use the distances to find the nearest neighbor distance for each point.

This function gives us the average distance for the k nearest other points, as well as the indicies of those points. We can then sort the values and plot them to see what a good epsilon value might be. The best epsilon value is the point where the average distance to the nearest neighbors starts to increase, or the elbow. This represents the point where the density starts to decrease, as each subsequently farther point starts to get much farther away. We pick an epsilon around that curve to "target" the point at which the neighbors stop being close enough to be considered part of the same cluster.

```python
## Elbow method

from sklearn.neighbors import NearestNeighbors

neigh = NearestNeighbors(n_neighbors=5, )
nbrs = neigh.fit(X)
distances, indices = nbrs.kneighbors(X)

# Plotting K-distance Graph
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.figure(figsize=(20,10))
plt.plot(distances)
plt.title('K-distance Graph',fontsize=20)
plt.xlabel('Data Points sorted by distance',fontsize=14)
plt.ylabel('Epsilon',fontsize=14)
plt.show()
```

```python
db = DBSCAN(eps=0.15, min_samples=10)
db.fit(X)

y_pred = db.fit_predict(X)
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1],c=y_pred, cmap='Paired')
plt.title("Clusters determined by DBSCAN")
```

The predictions are the cluster labels for each point. We can plot the results in a countplot to see how the clusters look. The labels of -1 are the noise points, those that don't belong to any cluster.

```python
sns.countplot(x=y_pred)
```

#### DBSCAN Results

DBSCAN is useful because it can work without having to specify the clusters beforehand, and it can handle data with some noise, which is often the case in real-world data. On the downside, setting the hyperparameters can be tricky, and if we have clusters that have inconsistent densities, then DBSCAN may not work well as it can't adapt well to the data.

We can apply this to a real dataset, without many differences from the other clustering algorithms. We'll project down to 2D for visualization purposes.
